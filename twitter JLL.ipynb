```python
postings = soup.find_all('div', {"class":"css-901oao r-1nao33i r-37j5jr r-a023e6 r-16dba41 r-rjixqe r-bcqeeo r-bnwqim r-qvutc0"})

tweets = []
while True:
    for posts in postings:
        tweets.append(posts.text)
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    time.sleep(3)
    soup = BeautifulSoup(driver.page_source, 'lxml')    
    postings = soup.find_all('div', {"class":"css-901oao r-1nao33i r-37j5jr r-a023e6 r-16dba41 r-rjixqe r-bcqeeo r-bnwqim r-qvutc0"})
    tweets2 = list(set(tweets)) # to remove the duplicates
    if len(tweets)> 200:
        break
```

```python
jll_df = pd.DataFrame(tweets2)
jll_df
```
